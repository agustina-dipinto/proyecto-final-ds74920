# **Descripción del Dataset**

El Dataset presenta información proporcionada por 20000 personas sobre diversos factores que afectan la calidad del sueño y la salud visual, incluyendo duración del sueño, niveles de estrés, actividad física practicada, historial médico, y uso del celular antes de dormir. Entender la relación entre la calidad del sueño, los niveles de estrés, y la exposición a las pantallas puede ser de ayuda al analizar cómo el estilo de vida moderno impacta en la salud.

**Descripción de columnas**
* Gender: género del individuo (M/F) – dato cualitativo
* Age: edad del individuo en años – dato cuantitativo
* Sleep duration: promedio de la duración del sueño en horas – dato cuantitativo
* Sleep quality: calidad del sueño reportada (Escala: 1- Pobre a 5- Excelente) – dato cualitativo
* Stress level: nivel de estrés (Escala: 1- Bajo a 5- Alto) – dato cualitativo
* Blood pressure: Presión arterial (Sistólico / diastólico en mmHg) – dato cuantitativo
* Heart rate: frecuencia cardíaca (lpm= latidos por minuto) – dato cuantitativo
* Daily steps: pasos diarios promedios – dato cuantitativo
* Physical activity: minutos de actividad física por día – dato cuantitativo
* Height: altura en cm – dato cuantitativo
* Weight: peso en kg- dato cuantitativo
* Sleep disorder: trastorno del sueño (Y/N) - dato cualitativo
* Wake up during night: sueño interrumpido (Y/N) - dato cualitativo
* Smoker: fumador (Y/N) - dato cualitativo
* Caffeine consumption: consumo de cafeína (Y/N) - dato cualitativo
* Alcohol consumption: consumo de alcohol (Y/N) - dato cualitativo
*	Medical issues: condición médica existente (Y/N) - dato cualitativo
*	Ongoing medication: medicación en curso (Y/N) - dato cualitativo
*	Smart device before sleep: uso del celular antes de dormir (Y/N)
*	Average screentime: tiempo de pantalla diario promedio en horas – dato cuantitativo
*	Blue light filter: uso de filtro luz azul (Y/N) – dato cualitativo
*	Discomfort Eye-strain: presencia de fatiga visual (Y/N) – dato cualitativo
* Redness in eye: ojos rojos (Y/N) – dato cualitativo
*	Irritation in eye: irritación o picazón en los ojos (Y/N) -dato cualitativo
*	Dry eye disease: diagnóstico o síntomas de enfermedad del ojo seco (Y/N) – dato cualitativo

**Objetivos**
1. Explorar la relación entre la cantidad y calidad del sueño con el género, edad y estilo de vida (estrés, actividad física, consumo de sustancias, uso de pantallas y estado de salud).
2. Analizar el impacto de la exposición a pantallas y el uso de filtro de luz azul en la salud ocular y la calidad del sueño.
3. Investigar la conexión entre los niveles de estrés y el bienestar general (sueño, salud física y ocular).

**Hipótesis**
1. Las personas con menos estrés que practican más actividad física  tienen mejor calidad y mayor cantidad de sueño.
2. Las mujeres y las personas mayores duermen menos horas y de peor calidad.
3. La menor exposición a pantallas y/o el uso de filtro de luz azul reduce la fatiga y la irritación ocular.
4. Las personas con bajo estrés tienen mejor salud en general, con menos trastornos del sueño y problemas oculares.

## **Hipótesis 1:**
Las personas con menos estrés que practican más actividad física tienen mejor calidad y mayor cantidad de sueño.

```python
df["Stress level"].value_counts()
```

```python
plt.figure(figsize=(8, 5))

sns.countplot(x="Stress level", data=df, palette="coolwarm")

plt.title("Distribución de Frecuencia del Nivel de Estrés", fontsize=14)
plt.xlabel("Nivel de Estrés (1 = Bajo, 5 = Alto)", fontsize=12)
plt.ylabel("Cantidad de Personas", fontsize=12)
```

Se observa una distribución de datos uniforme en cuanto a los niveles de estrés.

## **Tendencia horas de sueño vs. niveles de estrés**

```python
df_sorted = df.sort_values(by="Stress level")

# Aplicao rolling mean con ventana de 5 valores
df_sorted["Sleep Rolling Mean"] = df_sorted["Sleep duration"].rolling(window=5).mean()

plt.figure(figsize=(8, 5))
sns.lineplot(x=df_sorted["Stress level"], y=df_sorted["Sleep Rolling Mean"], marker="o", color="darkred")

plt.title("Tendencia de Horas de Sueño según Nivel de Estrés", fontsize=14)
plt.xlabel("Nivel de Estrés", fontsize=12)
plt.ylabel("Media Móvil de Horas de Sueño", fontsize=12)
plt.grid(True)

plt.show()
```

## **Actividad física vs. Calidad de sueño y estrés**

```python
# Agrupao los datos por actividad física y calcular promedios
df_grouped = df.groupby(pd.cut(df["Physical activity"], bins=[0, 30, 60, 90, 120, np.inf]))[["Sleep quality", "Stress level"]].mean()

plt.figure(figsize=(10, 6))
sns.heatmap(df_grouped[["Sleep quality", "Stress level"]],
            cmap="coolwarm", annot=True, fmt=".2f", linewidths=0.5,
            cbar_kws={'label': 'Valor Promedio'})

plt.title("Actividad Física vs. Calidad del Sueño y Estrés", fontsize=16, fontweight="bold")
plt.xlabel("Variable Analizada", fontsize=14)
plt.ylabel("Nivel de Actividad Física", fontsize=14)
plt.xticks(rotation=0)
plt.yticks(rotation=0)

plt.show()
```

En cuanto a la hipótesis 1 se podría decir que las primeras visualizaciones indican la posibilidad de su confirmación.

##**Hipótesis 2**
Las mujeres y las personas mayores duermen menos horas y de peor calidad.

```python
plt.figure(figsize=(7, 5))
sns.boxplot(x="Gender", y="Sleep duration", data=df, palette="pastel")

plt.title("Distribución de Horas de Sueño según Género", fontsize=14)
plt.xlabel("Género", fontsize=12)
plt.ylabel("Horas de Sueño", fontsize=12)
plt.show()
```

```python
df_age = df.groupby("Age")["Sleep quality"].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.lineplot(x="Age", y="Sleep quality", data=df_age, marker="o", color="red")

plt.title("Relación entre Edad y Calidad del Sueño", fontsize=14)
plt.xlabel("Edad", fontsize=12)
plt.ylabel("Calidad del Sueño (1-5)", fontsize=12)
plt.show()
```

Las visualizaciones no respaldan la hipótesis 2 ya que no se observan diferencias en el sueño según el genero y tampoco se nota una tendencia clara según la edad.

## **Hipótesis 3:**
La menor exposición a pantallas y/o el uso de filtro de luz azul reduce la fatiga y la irritación ocular

```python
df['Blue-light filter'] = df['Blue-light filter'].astype('category')
df['Discomfort Eye-strain'] = df['Discomfort Eye-strain'].astype('category')
df['Redness in eye'] = df['Redness in eye'].astype('category')
df['Itchiness/Irritation in eye'] = df['Itchiness/Irritation in eye'].astype('category')
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Blue-light filter', hue='Discomfort Eye-strain')
plt.title('Efecto del Uso de Filtro de Luz Azul en la Fatiga Ocular')
plt.xlabel('Uso de Filtro de Luz Azul')
plt.ylabel('Número de Personas')
plt.legend(title='Fatiga Ocular')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Average screen time', y='Itchiness/Irritation in eye')
plt.title('Relación entre el Tiempo de Pantalla Promedio y la Irritación Ocular')
plt.xlabel('Tiempo Promedio de Pantalla (horas)')
plt.ylabel('Irritación Ocular')
plt.show()
```

No se observa en las visualizaciónes respaldo a la hipótesis.

# **Hipótesis 4:**
Las personas con bajo estrés tienen mejor salud en general, con menos trastornos del sueño y problemas oculares.

```python
g1 = sns.FacetGrid(df, col='Stress level', col_wrap=3, height=4)
g1.map(sns.countplot, 'Sleep disorder', order=df['Sleep disorder'].value_counts().index)
g1.set_titles(col_template='Nivel de Estrés: {col_name}')
g1.set_axis_labels('Trastornos del Sueño', 'Número de Personas')
g1.fig.suptitle('Relación entre el Nivel de Estrés y los Trastornos del Sueño', fontsize=16, y=1.05)

plt.show()
```

```python
g2 = sns.FacetGrid(df, col='Stress level', col_wrap=3, height=4)
g2.map(sns.countplot, 'Discomfort Eye-strain', order=df['Discomfort Eye-strain'].value_counts().index)
g2.set_titles(col_template='Nivel de Estrés: {col_name}')
g2.set_axis_labels('Fatiga Ocular', 'Número de Personas')
g2.fig.suptitle('Relación entre el Nivel de Estrés y la Fatiga Ocular', fontsize=16, y=1.05)

plt.show()
```

Tampoco se observa en los datos la comprobación de la hipótesis
